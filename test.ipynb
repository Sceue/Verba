```python
import requests
import json

# The URL of the API endpoint
url = 'http://localhost:8000/api/generate'  # Change this to the actual URL of your API

# The data payload for the request
data = {
    "query": "Who is the director of msaii program?",
    "context": "",
    "conversation": []
}

# Headers to indicate that the payload is JSON
headers = {'Content-Type': 'application/json'}

# Sending the POST request
response = requests.post(url, data=json.dumps(data), headers=headers)

# Checking if the request was successful
if response.status_code == 200:
    # Parsing the response JSON and printing the result
    result = response.json()
    print("Generated Response:", result)
else:
    print("Error:", response.status_code, response.text)
```

```python
import requests

def query_for_document(query):
    """Retrieve document based on query."""
    url = 'http://localhost:8000/api/query'  # Adjust as necessary
    payload = {"query": query}
    response = requests.post(url, json=payload)
    if response.status_code == 200:
        data = response.json()
        # Assuming the response includes document details; adjust based on your actual response structure
        if data['documents']:
            # Returning the first document's details as an example
            return data['documents'][0]
        else:
            print("No documents found for the query.")
            return None
    else:
        print(f"Failed to query document: {response.text}")
        return None

def generate_answer_with_context(query, context):
    """Generate an answer using the retrieved document context."""
    url = 'http://localhost:8000/api/generate'  # Adjust as necessary
    payload = {
        "query": query,
        "context": context,
        "conversation": []  # Modify as per your requirements
    }
    response = requests.post(url, json=payload)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to generate answer: {response.text}")
        return None
```

```python
# Example query
query = "What are the core courses for MSAI?"

# Step 1: Retrieve relevant document/context
document = query_for_document(query)
if document:
    # Assuming 'text' contains the document's context; adjust as per your actual data structure
    context = document.get('text', '')
    
    # Step 2: Generate answer using the context
    answer = generate_answer_with_context(query, context)
    if answer:
        print("Generated Answer:", answer.get('system'))  # Adjust based on your response structure
    else:
        print("Failed to generate an answer.")
else:
    print("Document retrieval failed.")
```

```python
import requests

def generate_answer(query, context):
    """Generate an answer using the specified query and context."""
    url = 'http://localhost:8000/api/generate' 
    payload = {
        "query": query,
        "context": context,
        "conversation": []  
    }
    response = requests.post(url, json=payload)
    if response.status_code == 200:
        data = response.json()
        print(f"Generated Answer: {data['system']}")
        return data['system']
    else:
        print(f"Error generating answer: {response.status_code}")
        return None

query = "Who is the first president of CMU?"
context = ""  
answer = generate_answer(query, context)
print("Answer:", answer)
```

```python
import requests

def query_retrieve_document_and_generate_answer(query):
    """Query to retrieve relevant chunks, get document details, and generate an answer."""
    query_url = 'http://localhost:8000/api/query'
    document_url = 'http://localhost:8000/api/get_document'
    generate_url = 'http://localhost:8000/api/generate'
    
    query_payload = {"query": query}
    query_response = requests.post(query_url, json=query_payload)
    if query_response.status_code == 200:
        query_data = query_response.json()
        documents = query_data.get("documents", [])
        if not documents:
            print("No relevant documents found for the query.")
            return None
        context = query_data.get("context", "")
        print(f"Context: {context}")
        
        first_document_id = documents[0]["doc_uuid"]
        document_payload = {"document_id": first_document_id}
        document_response = requests.post(document_url, json=document_payload)
        
        if document_response.status_code == 200:
            document_data = document_response.json()
            # print(f"Document Details: {document_data}, Keys: {list(document_data.keys())}")
            # document_text = document_data["document"]["properties"]["text"] 

            
            generate_payload = {
                "query": query,
                "context": context,  
                "conversation": []
            }
            generate_response = requests.post(generate_url, json=generate_payload)
            
            if generate_response.status_code == 200:
                generate_data = generate_response.json()
                print(f"Generated Answer: {generate_data['system']}")
                return generate_data['system']
            else:
                print(f"Error generating answer: {generate_response.status_code}")
                return None
        else:
            print(f"Error retrieving document details: {document_response.status_code}")
            return None
    else:
        print(f"Error querying for documents: {query_response.status_code}")
        return None

# Example usage
query = "Who is the first president of CMU?"
answer = query_retrieve_document_and_generate_answer(query)
print("Answer:", answer)
```
